### 0.  using %

```python
questions = ['name', 'quest', 'favorite color']
answers = ['lancelot', 'the holy grail', 'blue']
for q, a in zip(questions, answers):
    #print('What is your {0}?  It is {1}.'.format(q, a)) 
    print("What is your %s ? \t It is %s." %(q, a))
```

### 3. unpack a tuple

```python
library=[('name','topic','pages'),
         ('AAAA','SDASDSADSADAS','601'),
         ('BBBB','DDDDDDDDD','21'),
         ('CCCC','ADCVFW','144')]
```

for book in library:
    print(book)

```python
for book in library:
    print(f'{book}')
```

```python
for book in library:
    print(f'author is {book[0]}')
```

```python
for author, topic, pages in library:
    print(f'{author}{topic}{pages}')
```

### add some spaces

```python
for author, topic, pages in library:
    print(f'{author:{10}}{topic:{30}}{pages:{10}}')
```

### add different aline

```python
for author, topic, pages in library:
    print(f'{author:{10}}{topic:{30}}{pages:>{10}}')
```

### or we can fill the aline with specific charaters

```python
for author, topic, pages in library:
    print(f'{author:{10}}{topic:{30}}{pages:.>{10}}')
```
