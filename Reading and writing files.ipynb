```python
myfile = open('open.txt')
print(myfile)
```

### for loop 

```python
for line in myfile:
    print(line)
```

### read()

```python
# cursor rached end of this file. So you cannot read it again
myfile.read()
```

```python
myfile.seek(0) # reset the cursor
```

### Introducing  Write a file

```python
# w+ means everything in original file is overwriten and deleted
myfile = open('test.txt','w+') 
```

```python
myfile.read() # a empty file
```

```python
myfile.write('This is the second version of test.txt')
```

```python
myfile.seek(0)
myfile.read()
```

```python
myfile.close()
```

### Introducing  append a file

```python
myfile = open('whoops.txt','a+')
```

```python
myfile.write('appending the first line')
```

```python
myfile.seek(0)
myfile.readlines()
```

```python
myfile.write(' still appending to the first line')
```

```python
myfile.seek(0)
myfile.readlines()
```

```python
myfile.write('\nappending the second line')
```

```python
myfile.seek(0)
myfile.readlines()
```

```python
myfile.seek(0)
print(myfile.read())
```

```python
myfile.close()
```

### with open mehtod can make sure when you finish your code, it will close the file automaticly

```python
with open('whoops.txt','r') as mynewfile:
    lines = mynewfile.readlines()
```
